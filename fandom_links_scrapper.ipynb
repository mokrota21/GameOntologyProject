```python
import pandas as pd
```

```python
df = pd.read_csv('metacritic_2010_2024.csv')
df
```

```python
import re 

def url_fandom(game_name):
    return "https://" + re.sub('[^0-9a-zA-Z]+', '', game_name)[:63] + ".fandom.com/wiki/"
```

```python
from bs4 import BeautifulSoup
import requests

def get_title_fandom(game_name):
    url = url_fandom(game_name)
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    title = soup.title.string
    return title

def exist_fandom(game_name):
    title = get_title_fandom(game_name=game_name)
    return title != 'Not a valid community |  | Fandom'

def exist_en_fandom(game_name):
    try:
        title = get_title_fandom(game_name=game_name)
        return title != 'Wiki Index |  | Fandom'
    except:
        return False
```

```python
def make_url_column(game_title):
    if exist_fandom(game_title):
        try:
            url = url_fandom(game_title)
            print(f"Game {game_title} has fandom page! {url}")
            return url
        except Exception as e:
            print(f"Error with {game_title}: {e}")
    return None
```

```python
df['fandom_url'] = df['Title'].apply(make_url_column)
```

```python
df
```

```python
urls_df = df[['fandom_url', 'Title']]
```

```python
urls_df
```

```python
urls_df = urls_df.dropna().reset_index().drop('index', axis=1)
urls_df.to_csv('fandom_links.csv')
```

```python
english_urls = urls_df
english_urls['english?'] = english_urls['Title'].apply(exist_en_fandom)
```

```python
non_english_urls = english_urls
```

```python
english_urls = english_urls[english_urls['english?'] == True].reset_index().drop(['index', 'english?'], axis=1)
```

```python
english_urls.to_csv('fandom_en_links.csv')
```

```python
non_english_urls[non_english_urls['english?'] == False].reset_index().drop(['index', 'english?'], axis=1).to_csv('fandom_non_en_links.csv')
```
